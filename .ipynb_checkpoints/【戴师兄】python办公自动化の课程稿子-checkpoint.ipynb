```python
from docx import Document
from docx.shared import Inches

document = Document()

document.add_heading('Document Title', 0)

p = document.add_paragraph('A plain paragraph having some ')
p.add_run('bold').bold = True
p.add_run(' and some ')
p.add_run('italic.').italic = True

document.add_heading('Heading, level 1', level=1)
document.add_paragraph('Intense quote', style='Intense Quote')

document.save('demo.docx')
```

```python
import pandas as pd
```

```python
video_list = pd.read_excel('video_list.xlsx')
speech_text = pd.read_excel('speech_text.xlsx')
```

```python
video_list['AwemeId'] = video_list['AwemeId'].astype(str)
speech_text['VideoId'] = speech_text['VideoId'].astype(str)
```

```python
merge = pd.merge(video_list, speech_text, how='inner', left_on='AwemeId', right_on='VideoId')
merge.head(20)
```

```python
document = Document()
for i in range(len(merge)):
    print(merge.iloc[i][''])
    document.add_heading(merge.iloc[i][''], level=1)
    document.add_heading(merge.iloc[i][''], level=2)
    document.add_paragraph(merge.iloc[i][''])
document.save('demo.docx')
```

## 

```python
document = Document()
for i in range(len(merge)):
    if merge.iloc[i][''] != merge.iloc[i-1][''] or i == 0:
        document.add_heading(merge.iloc[i][''], level=1)
        
    document.add_heading(merge.iloc[i][''], level=2)
    document.add_paragraph(f'{merge.iloc[i]["BloggerName"]}')
    document.add_paragraph(f'douyin.com/video/{merge.iloc[i]["AwemeId"]}')
    document.add_paragraph(merge.iloc[i][''])
document.save('demo.docx')
```
